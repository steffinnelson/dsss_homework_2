```
# Function to run OCR on an image region
def extract_text_from_region(image, bbox):
    x1, y1, x2, y2 = map(int, bbox)
    roi = image[y1:y2, x1:x2]  # Extract region of interest
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    text = pytesseract.image_to_string(gray, config='--psm 6').strip()  # Extract text
    return text
```

```
# Define component classes for rule-based classification
MCB_CLASSES = ["1 Pole Circuit Breaker", "2 Pole Circuit Breaker", "3 Pole Circuit Breaker", "4 Pole Circuit Breaker"]
CONTACTOR_CLASSES = ["1 pole contactor", "4 pole contactor"]
OTHER_CLASSES = ["DIN Rail", "Surge protection Device", "Terminal Block"]
```

```
# Rule-based classification function
def classify_component(yolo_label, ocr_text):
    ocr_text = ocr_text.lower()

    # If OCR detects pole numbers, override YOLO prediction
    if "1p" in ocr_text:
        return "1 Pole Circuit Breaker"
    elif "2p" in ocr_text:
        return "2 Pole Circuit Breaker"
    elif "3p" in ocr_text:
        return "3 Pole Circuit Breaker"
    elif "4p" in ocr_text:
        return "4 Pole Circuit Breaker"

    # If OCR detects contactor-related words
    if "contactor" in ocr_text:
        if "1p" in ocr_text:
            return "1 pole contactor"
        elif "4p" in ocr_text:
            return "4 pole contactor"

    # If OCR suggests "DIN Rail"
    if "din" in ocr_text:
        return "DIN Rail"

    # If OCR suggests "Surge Protection Device"
    if "spd" in ocr_text or "surge" in ocr_text:
        return "Surge protection Device"

    # If OCR suggests "Terminal Block"
    if "terminal" in ocr_text:
        return "Terminal Block"

    # If no clear OCR classification, return YOLO prediction
    return yolo_label
```

```
# Function to annotate image
def annotate_image(image, results):
    box_annotator = sv.BoxAnnotator()
    label_annotator = sv.LabelAnnotator(text_color=sv.Color.BLACK)

    detections = sv.Detections.from_ultralytics(results)
    annotated_image = box_annotator.annotate(image.copy(), detections=detections)
    annotated_image = label_annotator.annotate(annotated_image, detections=detections)

    return annotated_image
```

```
# Function to evaluate bounding box IoU
def compute_iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])

    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou
```

```
import os
import cv2
from ultralytics import YOLO

# Load YOLO model
model = YOLO("/content/runs/detect/train2/weights/best.pt")  # Change to your trained model

def extract_text_from_region(image, box):
    """ Extracts OCR text from a detected region (Placeholder). """
    return "SampleText"

def classify_component(yolo_label, ocr_text):
    """ Applies rule-based classification based on OCR and YOLO label. """
    return f"Corrected_{yolo_label}"

def convert_to_yolo_format(box, img_width, img_height):
    """ Converts absolute bounding box (xyxy) to YOLO format. """
    x1, y1, x2, y2 = box
    x_center = (x1 + x2) / 2 / img_width
    y_center = (y1 + y2) / 2 / img_height
    width = (x2 - x1) / img_width
    height = (y2 - y1) / img_height
    return x_center, y_center, width, height

def save_corrected_labels(image_name, detections, output_folder="/content/corrected_labels"):
    """ Saves corrected labels in YOLO format """
    os.makedirs(output_folder, exist_ok=True)  # Ensure folder exists

    # Convert image extension to .txt for label file
    label_filename = os.path.join(output_folder, os.path.splitext(image_name)[0] + ".txt")

    with open(label_filename, 'w') as label_file:
        for det in detections:
            class_name = det["final_label"]
            box = det["box"]

            # Check if class exists in model.names
            if class_name in model.names.values():
                class_id = list(model.names.keys())[list(model.names.values()).index(class_name)]
            else:
                print(f" Warning: Class '{class_name}' not found in YOLO classes. Skipping...")
                continue

            # Normalize bounding box
            yolo_box = convert_to_yolo_format(box, 1024, 1024)  # Change 1024x1024 based on real size

            # Write to file in YOLO format
            label_file.write(f"{class_id} {yolo_box[0]:.6f} {yolo_box[1]:.6f} {yolo_box[2]:.6f} {yolo_box[3]:.6f}\n")

    print(f" Saved corrected labels: {label_filename}")

def annotate_image(image, results):
    """ Annotates image with bounding boxes and labels """
    for box, cls in zip(results.boxes.xyxy, results.boxes.cls):
        cv2.rectangle(image, (int(box[0]), int(box[1])), (int(box[2]), int(box[3])), (0, 255, 0), 2)
        cv2.putText(image, model.names[int(cls)], (int(box[0]), int(box[1]) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return image

def process_image(image_path, output_label_folder="/content/corrected_labels"):
    """ Pipeline to run detection, OCR, and classification """
    image = cv2.imread(image_path)
    img_height, img_width = image.shape[:2]

    results = model.predict(image, conf=0.25)[0]
    final_detections = []

    for box, cls, conf in zip(results.boxes.xyxy, results.boxes.cls, results.boxes.conf):
        yolo_label = model.names[int(cls)]
        ocr_text = extract_text_from_region(image, box)

        # Apply rule-based classification
        final_label = classify_component(yolo_label, ocr_text)

        # Validate class ID
        if final_label in model.names.values():
            final_label_id = list(model.names.keys())[list(model.names.values()).index(final_label)]
        else:
            print(f" Skipping unknown class: {final_label}")
            continue

        # Convert bbox to YOLO format
        yolo_box = convert_to_yolo_format(box.tolist(), img_width, img_height)

        # Store corrected label
        final_detections.append({
            "box": box.tolist(),
            "yolo_label": yolo_label,
            "ocr_text": ocr_text,
            "final_label": final_label,
            "confidence": float(conf)
        })

    # Save corrected labels
    save_corrected_labels(os.path.basename(image_path), final_detections, output_label_folder)

    # Annotate final image
    annotated_image = annotate_image(image, results)

    return final_detections, annotated_image

###  Process All Images in the Folder ###
input_images_folder = "/content/control-cabinet-diassembly-1/train/images"  # UPDATE THIS
output_labels_folder = "/content/corrected_labels"

# Ensure output folder exists
os.makedirs(output_labels_folder, exist_ok=True)

# Process all images
image_files = [f for f in os.listdir(input_images_folder) if f.endswith((".jpg", ".png"))]

for i, image_name in enumerate(image_files):
    image_path = os.path.join(input_images_folder, image_name)
    print(f" Processing ({i+1}/{len(image_files)}): {image_name}")
    detections, annotated_img = process_image(image_path, output_labels_folder)

print(" All images processed with rule-based corrections!")
```

```
!wget https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8n.pt

import torch
import torch.nn as nn
import torchvision.transforms as transforms
import easyocr
import cv2
import os
import numpy as np
from ultralytics import YOLO
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel

# Load BERT tokenizer and model for text encoding
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
text_encoder = BertModel.from_pretrained("bert-base-uncased")

# Define OCR reader
reader = easyocr.Reader(['en'])

# Custom Dataset class to integrate OCR
class YOLODataset(Dataset):
    def __init__(self, image_folder, label_folder, transform=None):
        self.image_folder = image_folder
        self.label_folder = label_folder
        self.image_files = os.listdir(image_folder)
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        image_path = os.path.join(self.image_folder, self.image_files[idx])
        label_path = os.path.join(self.label_folder, self.image_files[idx].replace('.jpg', '.txt'))

        # Load image
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        if self.transform:
            image = self.transform(image)

        # Read labels
        labels = []
        with open(label_path, 'r') as f:
            for line in f.readlines():
                labels.append(list(map(float, line.strip().split())))
        labels = torch.tensor(labels)

        # Perform OCR
        results = reader.readtext(image_path)
        extracted_text = " ".join([res[1] for res in results])

        # Encode text with BERT
        text_tokens = tokenizer(extracted_text, return_tensors='pt', padding=True, truncation=True, max_length=20)
        text_embedding = text_encoder(**text_tokens).last_hidden_state.mean(dim=1)

        return image, labels, text_embedding

# Data augmentation and preprocessing
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((640, 640)),
    transforms.ToTensor()
])

# Load dataset
dataset = YOLODataset(image_folder='/content/control-cabinet-diassembly-1/train/images', label_folder='/content/control-cabinet-diassembly-1/train/label/train/labels', transform=transform)
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Load YOLO model
model = YOLO("yolov8n.pt")

# Modify YOLO model to accept text embeddings
class MultiModalYOLO(nn.Module):
    def __init__(self, yolo_model, text_dim=768):
        super(MultiModalYOLO, self).__init__()
        self.yolo = yolo_model.model
        self.text_fc = nn.Linear(text_dim, 256)  # Reduce BERT feature size
        self.fusion = nn.Linear(256 + 1024, 1024)  # Merge with YOLO features

    def forward(self, image, text_embedding):
        yolo_features = self.yolo.model(image)
        text_features = self.text_fc(text_embedding)
        fused_features = torch.cat((yolo_features, text_features), dim=1)
        return self.fusion(fused_features)

# Instantiate modified model
multi_modal_model = MultiModalYOLO(model)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(multi_modal_model.parameters(), lr=0.001)

# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    for images, labels, text_embeddings in dataloader:
        optimizer.zero_grad()
        outputs = multi_modal_model(images, text_embeddings)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Save trained model
torch.save(multi_modal_model.state_dict(), "yolo_multimodal.pth")
```
